```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Wine dataset
http://archive.ics.uci.edu/ml/datasets/Wine+Quality

```python
df_wine = pd.read_csv("winequality-white.csv", sep = ';')
df_wine.head()
```

```python
df_wine.describe()
```

- All features in this dataset is quantitative and continuous except quality because all of this features can be measure directly and the range value of feature are continuous (float number)
- Quality is qualitative and ordinal because it cant be observed directly and it represent the characteristic of wine?

```python
def get_mean(feature):
    return np.mean(feature)

def get_variance(feature):
    return np.var(feature)

cols = df_wine.columns

for col in cols:
    feature_val = df_wine[col].values
    print(f"Get statistic about {col} feature: ")
    print("Mean: ", get_mean(feature_val))
    print("Variance: ", get_variance(feature_val))
```

```python
def get_corr_matrix(df):
    corr = df.corr()
    return corr

def get_covar_matrix(df):
    mat = df.cov()
    return mat

def vis_heatmap_matrix(matrix, title_val):
    plt.figure(figsize=(15, 10))
    sns.heatmap(matrix, 
            xticklabels=matrix.columns.values,
            yticklabels=matrix.columns.values,
            annot=True)
    plt.title(title_val)
    plt.show()
```

```python
correlation_mat = get_corr_matrix(df_wine)
covariace_mat = get_covar_matrix(df_wine)

vis_heatmap_matrix(correlation_mat, "Correlation between features")
vis_heatmap_matrix(covariace_mat, "Covariance between features")
```

The most correlated features is: **density** and **residual sugar**

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.decomposition import PCA

def get_pca_of_data(data, comp):
    pca = PCA(n_components=comp)
    pca.fit(data)
    return pca

def norm_arr(arr):
    normed = (arr - arr.mean(axis=0)) / arr.std(axis=0)
    return normed
```

```python
data = df_wine.drop(["quality"], axis = 1).values
data = norm_arr(data)
```

```python
pca = get_pca_of_data(data, len(cols)-1)
print(pca.explained_variance_)
```

```python
print("Total variation explained by first two principle components: ", sum(pca.explained_variance_[:2]))
```

```python
def vis_pca_compvar(comp_var, num_feat):
    plt.figure(figsize=(15, 10))
    tot = sum(comp_var)
    var_exp = [(i / tot) for i in sorted(comp_var, reverse=True)]
    print(var_exp)
    cum_var_exp = np.cumsum(var_exp)

    # plot explained variances
    plt.bar(range(1,num_feat), var_exp, alpha=0.5,
            align='center', label='individual explained variance')
    plt.step(range(1,num_feat), cum_var_exp, where='mid',
             label='cumulative explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal component index')
    plt.legend(loc='best')
    plt.show()
```

```python
vis_pca_compvar(pca.explained_variance_, len(cols))
```

```python
pca_2comp = get_pca_of_data(data, 2)
```

```python
X_pca = pca_2comp.transform(data)
X_pca.shape
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(X_pca[:, 0], X_pca[:, 1],
            c=df_wine["quality"], edgecolor='none', alpha=0.5)
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

- From the figure above, we can see that with two principle component the individual classes are not well separated. Because when we use only 2 principle component, it can only capture approximately 45% variance of our dataset
- As we increase the number of principal components, the variance of our data can be captured more and more

# Iris dataset

```python
df_iris = pd.read_csv("iris.data", names=["sepal-length", "sepal-width", "petal-length", "petal-width", "classes"])
df_iris.head()
```

```python
df_iris.describe()
```

- Sepal length, sepal width, petal length and petal width are quantitative and continuous features
- Classes is qualitative features

```python
def get_mean(feature):
    return np.mean(feature)

def get_variance(feature):
    return np.var(feature)

df_iris_feat = df_iris.drop(["classes"], axis=1)
cols = df_iris_feat.columns

for col in cols:
    feature_val = df_iris_feat[col].values
    print(f"Get statistic about {col} feature: ")
    print("Mean: ", get_mean(feature_val))
    print("Variance: ", get_variance(feature_val))
```

```python
correlation_mat = get_corr_matrix(df_iris_feat)
covariace_mat = get_covar_matrix(df_iris_feat)

vis_heatmap_matrix(correlation_mat, "Correlation between features")
vis_heatmap_matrix(covariace_mat, "Covariance between features")
```

- The most correlated features: sepal-length and petal-length

```python
data = norm_arr(df_iris_feat.values)
```

```python
pca = get_pca_of_data(data, len(cols))
print(pca.explained_variance_)
```

```python
print("Total variation explained by first two principle components: ", sum(pca.explained_variance_[:2]))
```

```python
vis_pca_compvar(pca.explained_variance_, len(cols)+1)
```

```python
pca_2comp = get_pca_of_data(data, 2)
X_pca = pca_2comp.transform(data)
X_pca.shape
```

```python
class_name = df_iris["classes"].unique()
class_mapper = {k:v for k, v in zip(class_name, range(len(class_name)))}
```

```python
cls_idx = df_iris["classes"].map(class_mapper)
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(X_pca[:, 0], X_pca[:, 1],
            c=cls_idx, edgecolor='none', alpha=0.5)
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

From figure above, we can see that our individual classes are well separated.
