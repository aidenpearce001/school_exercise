## Signals

### plot the signal Input_1kHz_15kHz (in time domain)

```python
scale = np.arange(-np.pi, np.pi, 2*np.pi/len(Input_1kHz_15kHz))
plt.title('signal')
plt.plot(scale, Input_1kHz_15kHz)
plt.show()
```

## convert into frequency


```python
y = fft(Input_1kHz_15kHz)
```

## plot real/imaginary/magnitude/phase

```python
real = y.real
imaginary = y.imag
magnitude = abs(y)
phase = [cmath.phase(i) for i in y]
```

```python
scale1 = np.arange(-np.pi, np.pi, 2*np.pi/len(real))
plt.figure(figsize=(15, 4), dpi=80)
plt.subplot(1,4,1)
plt.title('real')
plt.plot(scale1, real)

scale2 = np.arange(-np.pi, np.pi, 2*np.pi/len(imaginary))
plt.subplot(1,4,2)
plt.title('imaginary')
plt.plot(scale2, imaginary)

scale3 = np.arange(-np.pi, np.pi, 2*np.pi/len(magnitude))
plt.subplot(1,4,3)
plt.title('magnitude')
plt.plot(scale3, magnitude)

scale4 = np.arange(-np.pi, np.pi, 2*np.pi/len(phase))
plt.subplot(1,4,4)
plt.title('phase')
plt.plot(scale4, phase)
```

## using Inverse-FFT to get back to time domain

```python
inverse = ifft(y)
```

```python
scale = np.arange(-np.pi, np.pi, 2*np.pi/len(inverse))
plt.plot(scale, inverse)
```
